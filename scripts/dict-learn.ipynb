```python
import torch
```

```python
d = 4
k = 4
n = 1

samples = []
representations = []

for i in range(k):
    samples.append(torch.randn(d))
    representations.append(torch.rand(n))

print(samples)
print(representations)

dictionary = torch.randn(d, n)


print(dictionary)
```

```python
recon = dictionary * representations[0]
print(recon)
print(recon.shape)
```
