```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/results_nobank.csv')



# Set the Model column as the index
df.set_index("Model", inplace=True)

# Transpose the DataFrame to make plotting easier
df = df.T

# Plot the data
plt.figure(figsize=(10, 6))
for model in df.columns:
    plt.plot(df.index, df[model], marker='o', label=model)

plt.xlabel('Dataset')
plt.ylabel('Score')
plt.title('Model Performance Across Different Datasets')
plt.legend(title='Model')
plt.grid(True)
plt.show()
```
