```python
from transformers import GPT2Model
import torch
from transformer_lens import HookedTransformer
from transformers import GPT2Tokenizer, GPT2LMHeadModel
import time

device = 'cuda'

g = HookedTransformer.from_pretrained("gpt2-small", device = device)
g2 = HookedTransformer.from_pretrained("gpt2", device = device)

tokenizer = GPT2Tokenizer.from_pretrained('gpt2')

# print the number of parameters in both models

print("Number of parameters in gpt2-small:", sum(p.numel() for p in g.parameters()))
print("Number of parameters in gpt2:", sum(p.numel() for p in g2.parameters()))
```

```python
tokenizer("asdfasdfasfadsfasdf asdfadsf fasd")
```

```python
batch_tokens = torch.zeros(1, 128).to(device).long()
```

```python
start = time.time()
for i in range(10):
    a, cache = g.run_with_cache(batch_tokens, attention_mask=[], prepend_bos=True, stop_at_layer=9)
end = time.time()

print(f"Time taken: {end - start}")
```
