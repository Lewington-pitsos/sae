```python
import torch
import torch.nn as nn
from transformers import GPT2Tokenizer, GPT2Model, GPT2LMHeadModel
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2')
tokenizer.pad_token = tokenizer.eos_token
```

```python
text = "one, two, three! four, five"

encoded_input = tokenizer(text, return_tensors='pt', padding=True)
output = model(**encoded_input)
```

```python
print(encoded_input.attention_mask)
print(encoded_input.input_ids.shape)
print(encoded_input.input_ids)
```

```python
past_key_values = None
max_length = 100
input_ids = encoded_input.input_ids
sequence = input_ids

for _ in range(max_length):
    # Get model outputs
    outputs = model(input_ids=input_ids, past_key_values=past_key_values)
    
    # Get the logits for the last token
    logits = outputs.logits[:, -1, :]
    
    # Get the predicted token (sampling, argmax, or another method)
    next_token = torch.argmax(logits, dim=-1).unsqueeze(-1)
    
    # Concatenate predicted token to input_ids
    sequence = torch.cat([sequence, next_token], dim=-1)
    input_ids = next_token

    # Update past_key_values for efficient decoding
    past_key_values = outputs.past_key_values
    
    # Check if the predicted token is the end-of-sequence token
    if next_token.item() == tokenizer.eos_token_id:
        break
```

```python
outputs = tokenizer.decode(sequence.squeeze(), skip_special_tokens=True)
print(outputs)
```
